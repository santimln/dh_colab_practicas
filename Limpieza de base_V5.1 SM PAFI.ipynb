### Gender

```python
# gender_x: elimino unknown y other

df = df[(df.gender_x != 'Other') & (df.gender_x != 'Unknown') ]
df.gender_x.value_counts()
```

```python
df.shape
```

### admissionheight & admissionweight --> IMC (indice de masa corporal)

```python
# admissionheight
# transformo valores en metros a centimetros
df = df.apply(lambda x: (x * 100) if x['admissionheight'] < 2.5 else x , axis=1)

#acoto rango de 40 a 250
df = df[(df.admissionheight > 40) & (df.admissionheight < 250) ]
```

```python
df.admissionheight.describe()
```

```python
#admissionweight
#acoto rango de 5 a 350
df = df[(df.admissionweight > 5) & (df.admissionweight < 350) ]
```

```python
df.admissionweight.describe()
```

```python
#Nueva columna IMC
df = df.assign(IMC = lambda x: x['admissionweight'] / ((x['admissionheight']/100)**2)) 
```

```python
df.shape
```

```python
df.IMC.describe()
```

```python
df = df[(df.IMC > 10) & (df.IMC < 80)]
```

```python
df.IMC.describe()
```

```python
### transformacion de variable continua IMC en categorica de acuerdo a definicion de obesidad 

bins_imc = [0, 18.5, 25, 30, 35, 40, 1000]
labels_imc = ["Peso bajo", "Normal", "Sobrepeso", "Obesidad grado I", "Obesidad grado II", "Obesidad grado III"]

# Obtengo una lista de intervalos
# right=False indice que el extremo derecho del intervalo no estÃ¡ incluÃ­do en la categorÃ­a
imc_categories = pd.cut(df.IMC, bins = bins_imc, labels = labels_imc, right=False)

print (imc_categories.value_counts())
```

```python
df["IMC"] = imc_categories
df.shape
```

### Nueva variable GCS (Glasgow coma score) a partir de meds, eyes, motor y verbal

```python
df.meds_x.value_counts()
```

```python
#Reemplazo nulos en eyes, motor, verbal por 0

df['eyes_x'] = df['eyes_x'].fillna(0)
df['motor_x'] = df['motor_x'].fillna(0)
df['verbal_x'] = df['verbal_x'].fillna(0)
```

```python
#Nueva columna GCS
df = df.assign(GCS = lambda x: x['eyes_x'] + x['motor_x'] + x['verbal_x']) 
```

```python
df.shape
```

```python
df.GCS.describe()
```

```python
### transformacion de variable GCS en categorica 

bins_gcs = [-1,9,13,16]
labels_gcs = ["Severo", "Moderado", "Leve"]

gcs_categories = pd.cut(df.GCS, bins = bins_gcs, labels = labels_gcs, right=False)

print (gcs_categories.value_counts())
```

```python
df["GCS"] = gcs_categories
df.shape
```
